```python
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import requests
import re
import pandas as pd
```

```python
url = "https://groww.in/recurring-deposit/rd-interest-rates"
req = requests.get(url)
soup = bs(req.content, "html.parser")
```

```python
for table in soup.find_all('table',{"dir":"ltr"}):
    table_data = table
```

```python
rows = []
for row in table_data.find_all('tr'):
    rows.append(row)
```

```python
bank_names = []
min_rate = []
max_rate = []
```

```python
for row in rows[1:]:
    data = []
    for row_data in row.find_all('td'):
        data.append(row_data)
    # print(data)
    bank_names.append(re.sub('[^A-Za-z\s+]','',data[0].text))
    min_rate.append(float(re.sub('[^0-9.\s+]','',data[1].text.split()[0])))
    max_rate.append(float(re.sub('[^0-9.\s+]','',data[1].text.split()[-1])))
```

```python
df = pd.DataFrame({"index":bank_names,"Bank name":bank_names,"Min rate":min_rate,"Max rate":max_rate})
df.sort_values(by='Min rate',inplace=True,ascending=False)
df.set_index('index',inplace=True)
df
```

```python
def getMaturity(bank,principle, timeInYears):
    rate = df.loc[bank,'Min rate']
    n = 4
    ans = 0
    progress = [principle]
    for i in range(1,(timeInYears*12)+1):
        ans = ans + (principle * pow((1 + (rate/(n*100))), n*(i/12)))
        progress.append(round(ans,2))
    return progress[-1]
```

```python
getMaturity('IDBI Bank',2000,10)
```
